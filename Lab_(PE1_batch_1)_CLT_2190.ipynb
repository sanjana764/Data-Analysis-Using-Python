```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

mean = 100
std_dev = 15
num_samples = 1000


data = np.random.normal(loc=mean, scale=std_dev, size=num_samples)
print(data)
```

```python
# Create a histogram
plt.hist(data, bins=30, density=True, alpha=0.6, label='Generated Data')
# Generate x-values for the theoretical curve
x = np.linspace(min(data), max(data), 100)

# Calculate the corresponding y-values using the probability density function (pdf)
y = (1 / (std_dev * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mean) / std_dev)**2)

# Plot the theoretical curve
plt.plot(x, y, color='red', linewidth=2, label='Theoretical Curve')
plt.xlabel('Values')
plt.ylabel('Probability Density')
plt.title('Normally Distributed Data')
plt.legend()
plt.show()
```
